```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its exact derivative
f = lambda x: np.e**(x)
df_exact = lambda x: np.e**(x)

x0 = 0
   # point of evaluation
hs = np.logspace(-1, 0, 50)  # step sizes
errors = []

for h in hs:
    df_approx = (f(x0 + h) - f(x0 - h)) / (2 * h)  # centered difference
    errors.append(abs(df_approx - df_exact(x0)))

plt.loglog(hs, errors, 'o-', label='Centered difference error')
plt.xlabel('Step size h')
plt.ylabel('Absolute error')
plt.title('Error vs Step Size (Centered Difference)')
plt.grid(True, which="both")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def forward_diff(f, x0, h):
    """Approximate derivative using the forward difference formula."""
    return (f(x0 + h) - f(x0)) / h

def backward_diff(f, x0, h):
    """Approximate derivative using the backward difference formula."""
    return (f(x0) - f(x0 - h)) / h

def centered_diff(f, x0, h):
    """Approximate derivative using the centered difference formula."""
    return (f(x0 + h) - f(x0 - h)) / (2 * h)


f = lambda x: np.e**x
df_exact = lambda x: np.e**x

x0 = 0 # point of evaluation
hs = np.logspace(-6, -1, 50) # step sizes from 10^-6 to 10^-1

errors_forward = []
errors_backward = []
errors_centered = []

for h in hs:
    df_f = forward_diff(f, x0, h)
    df_b = backward_diff(f, x0, h)
    df_c = centered_diff(f, x0, h)

    exact_val = df_exact(x0)
    errors_forward.append(abs(df_f - exact_val))
    errors_backward.append(abs(df_b - exact_val))
    errors_centered.append(abs(df_c - exact_val))

plt.figure(figsize=(8,6))
plt.loglog(hs, errors_forward, 'o-', label='Forward difference')
plt.loglog(hs, errors_backward, 's-', label='Backward difference')
plt.loglog(hs, errors_centered, '^-', label='Centered difference')

plt.xlabel('Step size h')
plt.ylabel('Absolute error')
plt.title('Error vs Step Size for Numerical Differentiation Methods')
plt.grid(True, which="both", ls='--')
plt.legend()
plt.show()
```
